Exploratory Data Analysis (EDA) on the cleaned UCI Air Quality dataset.

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Configs
sns.set(style="whitegrid")
```

```python
# Load cleaned dataset
df = pd.read_csv('E:/air_quality_forecasting/data/clean_air_quality.csv', parse_dates=['datetime'])
df.set_index('datetime', inplace=True)
df.head()
```

## Missing Values Analysis

```python
missing = df.isna().sum()
missing[missing > 0].sort_values(ascending=False)
```

## Time Series Plot of Pollutants

```python
df[['CO(GT)', 'NO2(GT)', 'NOx(GT)', 'C6H6(GT)']].plot(subplots=True, figsize=(15, 10), title='Pollutant Trends')
```

## Correlation Heatmap

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Between Variables')
```
